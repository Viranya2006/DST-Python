## Part 1 - Set Example

```python
fruit_set = {'apple', 'banana', 'orange', 'mango'}
print(fruit_set)
print('Length:', len(fruit_set))
for item in fruit_set:
    print(item)
```

## Part 2 - Tuple Example

```python
tupleA = ('banana', 'mango', 53, 12.5, False)
print(tupleA)
print(type(tupleA))
```

## Part 3 - Convert Tuple to List

```python
list_from_tuple = list(tupleA)
print(list_from_tuple)
print(type(list_from_tuple))
```

## Part 4 - Add Item to List

```python
list_from_tuple.append('ORANGE')
print(list_from_tuple)
```

## Part 5 - Convert List Back to Tuple

```python
tuple_from_list = tuple(list_from_tuple)
print(tuple_from_list)
```

## Part 7 - Dictionaries

## Part 8 - Dictionary Operations

```python
dictC = {
    'brand':'Porsche',
    'model':'911 turbo s',
    'year':2026,
    'colour':'miami blue'
}
print(dictC)
print(len(dictC))
print(dictC['colour'])
dictC.update({'year':2000})
print(dictC)
```

```python
dictC['Engine'] = '5000cc'
print(dictC)
```

```python
for key in dictC.keys():
    print(key)
```

```python
for key, value in dictC.items():
    print(key, value)
```
