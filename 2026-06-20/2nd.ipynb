## Handle missing values

```python
#remove rows that are completely null
cleaned_DF = orginal_DF.dropna(how="all")
```

```python
#Missing value percentage of the column
missing_percentage = ((cleaned_DF.isnull().sum()) / (len(cleaned_DF))) * 100
print(missing_percentage)
```

```python
print(missing_percentage[missing_percentage > 15])

# most of columns are in good state .If the percentage is greater that 65. should remove the colum
```
