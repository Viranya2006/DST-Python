## Import Revevent libraries

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('messy_data (1).csv')
print(df.shape)
```

## Handle Missing Values

```python
#Get overall messing count and percentage
missing_count = df.isnull().sum()
print(missing_count)
```

```python
#Get missing Percentage 
missing_count = (df.isnull().sum() / len (df)) * 100
print(missing_count)
```

```python
#Remove row that are completely null
print(df.shape)#before remove null columns
df_clean = df.dropna(how="all")
print(df_clean.shape)#after removing the null columns
```

```python
#Remove rows that were any of the critical columns are missing
df_clean = df.dropna (subset=['name','age','email'])
print(df_clean.shape)
```

## Imputation Methord

```python
df_clean['customer_rating'] = df_clean['customer_rating'].fillna(
    df_clean['customer_rating'].mode()[0]
)
```

## Handdle Inconsistance Categories

```python
print(f"Original Job Title Count: {df_clean['job_title'].nunique()}")
```

```python
print(df_clean['job_title'].value_counts())
```

```python
job_mapping ={
    'Sr. Developer' : 'Senior Developer',
    'senior developer' : 'Senior Developer',
    'Product Manager' : "Manager",
    'Regional Manager' : "Manager"
}
```

```python
df_clean['standarized_job'] = df_clean['job_title'].replace(job_mapping)
```

```python
print(df_clean['standarized_job'].value_counts)
```
