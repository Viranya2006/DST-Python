## Handle Missing Values

```python
# Total number of columns
total_cols = original_DF.shape[1]
original_DF['Null_percentage'] = ((original_DF.isnull().sum(axis=1))/total_cols)*100
print(original_DF.head(2))
```

```python
cleansed_DF = original_DF[original_DF['Null_percentage'] < 75].copy()
cleansed_DF.shape
```

## Handling Missing Values in Columns

```python
#Get overall column wise missing value count and percentage
missing_count = cleansed_DF.isnull().sum()
print(missing_count)
```

```python

missing_value = cleansed_DF.isnull().sum()
missing_percentage = (missing_count / len (cleansed_DF)) *100
print(missing_value)
print(missing_percentage)
```

```python
#display the coulumns with existing values

missing_data=pd.concat([missing_value,missing_percentage],axis=1,keys=['count','Percentage'])

print(missing_data)
```

```python
missing_data[missing_data['Percentage'] > 65].sort_values(by='Percentage', ascending=False)
```

```python
#Remove Columns
removed_columns = missing_data[missing_data['Percentage']>60].index
```

```python
#Drop the Columns
cleansed_DF = cleansed_DF.drop(columns=removed_columns)
cleansed_DF.shape
print("Removed columns count: "),len(removed_columns)
```

```python
# Count missing values
missing_count_a = cleansed_DF.isnull().sum()

# Calculate missing percentage
missing_percentage_a = (missing_count_a / len(cleansed_DF)) * 100

# Display columns with missing values
missing_data_a = pd.concat(
    [missing_count_a, missing_percentage_a],
    axis=1,
    keys=['count', 'Percentage']
)

print(missing_data_a)

missing_data_a[
    missing_data_a['Percentage'] > 0
].sort_values('Percentage', ascending=False)
```
